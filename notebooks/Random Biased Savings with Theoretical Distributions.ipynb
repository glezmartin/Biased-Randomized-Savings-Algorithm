## The idea
The savings heuristic is an algorithm that allows the establishment of acceptably good but suboptimal delivery routes. Its idea, as mentioned above, is based on establishing a list of the edges ordered according to the savings of joining them instead of making individual routes. But this solution is not optimal and it is possible to find better solutions.

For this purpose we can adopt an approach based on random search. Through random search we can eliminate the deterministic component of the algorithm and generate different solutions each time by choosing a different order of routes to consider from the list.

The idea of randomized search is based on the exploration of the solution space. The big disadvantage of performing a randomized search is that the meaning of the designed heuristic would be lost. For this reason, the biased random search is positioned as a good solution to this problem.

The random biased search allows to establish different probabilities of being chosen to the routes. In such a way that we can give preference to be chosen to those routes that would mean greater savings. In this way we would be giving to the random search the information generated by the heuristic savings. In this project we present two theoretical distributions to assign probabilities to the samples of the group of routes: the triangular distribution and the geometric distribution.

## Iterative search
Once implemented and understood the sense of the introduction of the random component in the heuristics. We must make a small additional change. It is elementary that if we allow the algorithm to have a stochastic behavior every time we run it to obtain a solution, this solution will be different.

To deal with this it is necessary to run several iterations so that at the end of this process we are left with the solution that has the lowest cost. Although this does not eliminate the stochastic component nor avoids the possibility that each time we run the algorithm we obtain a different solution, it provides robustness to the process and minimizes the range of possible results allowing us to obtain, approximately, solutions of a similar cost.

### Iterative triangular random biased search

```python
solution = random_biased_savings.iter_triangular_rand_biased_savings('P-n101-k4', iterations=50)
solution.plot_routes()
```

### Iterative geometric random biased search $\beta = 0.2$

```python
solution = random_biased_savings.iter_geometrical_rand_biased_savings('P-n101-k4', beta=0.2, iterations=50)
solution.plot_routes()
```
